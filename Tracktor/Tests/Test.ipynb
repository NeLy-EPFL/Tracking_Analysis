```python
for i in range(1,100):
    if ((i/3).is_integer()==True) and ((i/5).is_integer()==False):
        print("Fizz")
    if ((i/5).is_integer()==True) and ((i/3).is_integer()==False):
        print("Buzz")
    if ((i/3).is_integer()==True) and ((i/5).is_integer()==True):
        print("FizzBuzz")

    else:
        print(i)
```
